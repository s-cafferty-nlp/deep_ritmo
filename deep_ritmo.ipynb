### Demonstration of Deep Ritmo

```python
import pandas as pd
import spacy
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('processed_word_list.csv')
df.head()
```

```python
import deep_ritmo

epitran_model = 'spa-Latn'
spacy_model = spacy.load('es_dep_news_trf')
```

```python
P = deep_ritmo.Preprocess(epitran_model=epitran_model, spacy_model=spacy_model)
```

```python
from transformers import AutoTokenizer
tokenizer_checkpoint = 'caffsean/deep-ritmo-sampa-tokenizer'
tokenizer = AutoTokenizer.from_pretrained(tokenizer_checkpoint)


from transformers import BertForTokenClassification
model = BertForTokenClassification.from_pretrained('caffsean/bert-base-cased-deep-ritmo-sampa')
```

```python
R = deep_ritmo.DeepRitmo(tokenizer,model)
```

```python
word_list = list(df['original_text'])
```

```python
R.add_word_list(word_list)
```
